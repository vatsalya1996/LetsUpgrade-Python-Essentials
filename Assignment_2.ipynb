# For Loop to print prime numbers between 1 & 1000

```
for num in range (1, 1001):
 
    if (num > 1):
       for i in range (2, num):
           if (num % i == 0):
               break
       else:
           print(num)
```

 # Tables of 1 to 10 using Nested for loop

```
for i in range (1, 11):
    print("Multiplication table for ", i, ":")
    for j in range (1, 11):
        print(i, "X", j, "= ",i*j)
```
